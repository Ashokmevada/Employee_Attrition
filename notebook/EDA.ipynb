```python
category_columns = list(df.select_dtypes(include=['object']).columns)
numerical_columns  = list(df.select_dtypes(include=['int64']).columns)
category_columns , numerical_columns
```

```python
category_columns.remove("Attrition")
```

```python
# Calculate the number of rows and columns for the subplots grid
num_rows = len(category_columns)
num_cols = 2

# Create subplots based on the number of categorical columns
fig, axes = plt.subplots(num_rows, num_cols, figsize=(30, 50))

# Loop through each categorical column and create a bar plot
for i, column in enumerate(category_columns):
    ax1 = axes[i,0]
    ax2 = axes[i,1]
    df[column].value_counts().plot(kind='bar', ax=ax1)
    plt.tight_layout()
    ax1.set_title(f"{column} Distribution")
    ax1.set_xlabel(column)
    ax1.set_ylabel("Count")
    pd.crosstab(df['Attrition'] , df[column]).plot(kind= 'bar' , ax=ax2)
    ax2.set_title(f"Attritoin v/s {column}")
    ax2.set_xlabel("Attrition")
    ax2.set_ylabel(column)
    plt.tight_layout()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
